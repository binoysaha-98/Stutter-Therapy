```python
import pandas as pd
import os
from pandas import DataFrame

common_path = "/home/mansi/anaconda3/beproject/stutter_det/select1_collect"
print(common_path)
audio_segments = []
stutter = []
nonstutter = []
flag1 = 0
for subdir1, dirs1, files1 in os.walk(common_path):
    if(flag1 == 0):
        flag1 = 1        
        for subd1 in dirs1:
            #main audio folders
            print(subd1)
            flag2 = 0
            for subdir2, dirs2, files2 in os.walk(common_path + "/" + subd1):
                if(flag2 == 0):
                    flag2 = 1                    
                    for subd2 in dirs2:
                        #stutter or nonstutter folders
                        print(subd2)
                        flag3 = 0
                        for subdir3, dirs3, files3 in os.walk(common_path + "/" + subd1 + "/" + subd2):
                            if(flag3 == 0):
                                flag3 = 1
                                for filename in os.listdir(common_path + "/" + subd1 + "/" + subd2):
                                    #if os.path.isdir(common_path + "/" + subd1 + "/" + subd + "/" +filename) :
                                    if filename.endswith('.csv'):
                                        print(filename)
                                        audio_segments.append(subd1 + "/" + subd2 + "/" + filename)
                                        if subd2 == "nonstutter":
                                            nonstutter.append("1")
                                            stutter.append('0')
                                        else:
                                            nonstutter.append("0")
                                            stutter.append('1')

print(len(audio_segments))

dataset = {
    'audio_segment': audio_segments,
    'nonstutter': nonstutter,
    'stutter': stutter
}

df = DataFrame(dataset, columns= ['audio_segment', 'nonstutter', 'stutter'])

export_csv = df.to_csv(common_path + "/dataset_select1-collect.csv")

                
                            
                            
```
