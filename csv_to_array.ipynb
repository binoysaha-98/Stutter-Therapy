```python
import pandas as pd  
import numpy as np  
import math
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
audiodata = pd.read_csv("/home/mansi/anaconda3/beproject/stutter_det/select2_collect/dataset_select2-collect.csv")
print(len(audiodata))

F = audiodata.drop('nonstutter', axis=1) 
F = F.drop('stutter', axis=1)

G = audiodata.drop(audiodata.columns.values[1], axis=1)
G = G.drop(G.columns.values[0], axis=1)

Y = np.array(G)
print(Y.shape)
```

```python
i = 0
L = []
X = []

k=0

for index, row in F.iterrows():
    print(row['audio_segment'])
    if 'rep' in row['audio_segment']:
        seg_mfcc = pd.read_csv("/home/mansi/anaconda3/beproject/stutter_det/select2_collect/"+row['audio_segment'])
        z = np.array(seg_mfcc,dtype = np.float64)
        
        if z.shape[0]==13 and z.shape[1]==45:
            a = []
            for j in range(0,13):
                a.append(z[j][1:])
                print(len(a[0]))
            X.append(a)
            L.append(Y[k])
            i+=1
    k+=1
    
X = np.array(X)
L = np.array(L)

np.save('mfcc_select2_rep_13d.npy', X)
np.save('label_select2_rep_13d.npy', L)

print(L.shape)
print(X.shape)


```
